```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from CSV 
df = pd.read_csv('TGNNEXPLAINER-PUBLIC/benchmarks/results/featureImportance/feature_importance.csv')
df_sorted = df.sort_values('Importance Score', ascending=False)

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Event', y='Importance Score', data=df_sorted)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```
